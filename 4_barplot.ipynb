### Module importation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data reading

```python
data_dir = "data_stations/preprocessed_data/prec_peryear_perstate.csv"

data = pd.read_csv(data_dir)
```

### Cumulated precipitation plot

```python
categories = data['sg_estado'].unique()
years = data['ano'].unique()

fig, axs = plt.subplots(1, len(categories), figsize=(12, 5), sharey=True)
fig.suptitle('Precipitação Acumulada por Estado por Ano')

color_map = {'MG': 'blue', 'SP': 'orange', 'RJ': 'green', 'ES': 'red'}

for i, category in enumerate(categories):
    subset = data[data['sg_estado'] == category]
    
    axs[i].bar(subset['ano'],
               subset.groupby('ano')['soma_acumulada'].sum(), 
               color=[color_map[category]],
               edgecolor='black')
    
    axs[i].set_title(f'{category}')
    axs[i].set_xlabel('Ano')
    if i == 0:
        axs[i].set_ylabel('Precipitação acumulada')
        
    axs[i].set_xticks([2010, 2013, 2016, 2019, 2022])
    
plt.show()
```

### Mean precipitation plot

```python
categories = data['sg_estado'].unique()
years = data['ano'].unique()

fig, axs = plt.subplots(1, len(categories), figsize=(12, 5), sharey=True)
fig.suptitle('Precipitação Média por Estado por Ano')

color_map = {'MG': 'blue', 'SP': 'orange', 'RJ': 'green', 'ES': 'red'}

for i, category in enumerate(categories):
    subset = data[data['sg_estado'] == category]
    
    axs[i].bar(subset['ano'],
               subset.groupby('ano')['media'].sum(), 
               color=[color_map[category]],
               edgecolor='black')
    
    axs[i].set_title(f'{category}')
    axs[i].set_xlabel('Ano')
    if i == 0:
        axs[i].set_ylabel('Precipitação média')
        
    axs[i].set_xticks([2010, 2013, 2016, 2019, 2022])
    
plt.show()
```
